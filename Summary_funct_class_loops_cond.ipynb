1. Condicionales.

    1.1. Operadores de comparación en Python:
        
    <ul>
        <li>equal: <b>==</b></li>
        <li>not equal: <b>!=</b></li>
        <li>greater than: <b>></b></li>
        <li>less than: <b>&lt;</b></li>
        <li>greater than or equal to: <b>>=</b></li>
        <li>less than or equal to: <b>&lt;=</b></li>
    </ul>
    
    1.2. 'B'>'A' se considera True (se tiene en cuenta el orden alfabético para hacer la comparación).
    
    1.3. Sintaxis de if: (lo más importante es no olvidar los dos puntos)
    
    <ul>
        <li>if {condición}:
           <li>{consecuencias1}
        <li>elif {condición}:
           <li>{consecuencias2}
           <li>.
           <li>.
           <li>.
        <li>else:
           <li>{consecuenciasN}
    </ul>
        
    1.4. elif se escribe para que el nuevo if esté al mismo nivel que el if y el else, si escribimos if en su lugar, el bloque condicional del if anterior se cierra y se abre uno nuevo a partir de este if, que incluirá al else (en cuyo caso tenemos dos bloques condicionales, no uno).

2. Bucles

    
    2.1. La función range(n) te crea un vector de longitud n, que va de 0 hasta n-1. También puedes usar range (x,n) para que el vector empiece en x en lugar de en 0 (aunque seguirá llegando hasta n-1). Es muy útil para bucles for.
    
    __2.2. La sintaxis básica del bucle for es: for iteration_index_name in iteration_interval: ((indentación)) operaciones__. El intervalo de iteración puede ser una variable, en cuyo caso el índice de iteración será en cada iteración uno de los elementos de la variable, empezando por la izquierda.
    
    __2.3. Puedes asignar varios índices de iteración en el mismo for, para ello, el intervalo de iteración debe ser una matriz con ese nº de columnas(o filas, según se mire), de tal forma que cada índice itera sobre una de ellas.__
    
    __2.4. La sintaxis de un bucle while es: while ({condiciçon}): ((indentación)) operaciones.__
    
3. Funciones


    __3.1. Para definir una función escribimos: def function_name({argumentos de la función, separados por comas}): ((indentación)) operaciones.

    __3.2. Puedes usar la función return() para que la función te devuelva un output.__ También podrías usar el print, pero eso solo te valdría para imprimir el resultado, return da más juego. La función return también puede ser escrita sin los paréntesis.

    3.3. Si añades tres comillas (""") puedes añadir comentarios a la ayuda de tu nueva función (hay que escribir otras tres para cerrar este apartado dentro del código de la función.

    3.4. Si defines una variable dentro de una función, esa variable será una variable local, de modo que no sobreescribirá variables fuera del código de definición de la función.

    3.5. Una variable definida fuera de la función es una variable global y sí afecta a su definición.

    3.6. Escribir return(None) es equivalente a no escribirlo. Si almacenas el resultado de una función con return(None) en una variable, no te devuelve nada y su tipo es NoneType.

    __3.7. Puedes escribir operaciones dentro del return.

    __3.8. La función sum() suma los valores de una lista o tupla__. La función len(), vista anteriormente, también se usa sobre listas o tuplas. Ambas son ejemplos de funciones predefinidas en Python.

    __3.9. Puedes asignar valores por defecto en las funciones que crees, añadiendo un =default_value al argumento para el que quieras añadir un valor por defecto.__ Así, si no añades un valor para ese argumento en el paréntesis de la función, te devolverá el valor que se obtiene usando su valor por defecto.

    3.10. Por norma general, una variable local no puede ser llamada fuera del bloque donde la defines (en este caso, el bloque donde defines tu función). Sin embargo, puedes usar el comando global dentro de dicho bloque para convertir esa variable en global. Ejemplo: (...) global var_name (...). 

    3.11. La parte de un programa dentro de la cual una variable es accesible es llamada "Scope" de una variable.

4. Clases


    4.1. La sintaxis para crear una clase es la siguiente: class class_name(object): ((indentación)) {mínimo dos funciones: el constructor y el método}.
    
    4.2. Normalmente el constructor se nombra __init__ y en su argumento tiene una variable general y otras que están relacionadas con ella. Por ejemplo: def __init__(self, radius = 3, color = 'red'): ((indentación)) self.radius = radius (...). Obsérvese que se suelen usar valores por defecto al crear las funciones de las clases.
    
    4.3. Los métodos usan la misma variable general que el constructor y también pueden incluir otras nuevas.
    
    4.4. Una clase se usa como una función, solo que el argumento (object) se rellena con los argumentos del constructor (excepto la variable general, que no se incluye).
    
    4.5. La función dir() te da, al final, los métodos y variables de la clase.
    
    4.6. Puedes usar los nombres de las variables no generales de la clase como si fueran métodos para ver su valor una vez que los has almacenado en una variable. Por ejemplo, si almaceno un posible valor de la clase en x, x.var_name1 me da el valor que usé para la variable var_name1 en ese caso. También puedes usarlo para sobreescribir valores o con los métodos de la clase (en cuyo caso tienes que añadir () al final de su nombre).
