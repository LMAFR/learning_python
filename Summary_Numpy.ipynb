1. Numpy. Introducción y uso con arrays en 1D.

    1.1. Numpy es el equivalente a MATLAB en Python. Para acortar, se le suele dar el nickname np: **import numpy as np**.

    **1.2. Np trabaja con arrays, que son como las listas de Python pero que solo pueden contener un tipo de objeto por cada array.**

    **1.3. Podemos castear una lista a un array usando np.array({lista}).** Recordemos que las listas van entre corchetes. 

    1.4. La función range() nos genera una lista, así que es muy usada para crear arrays simples de forma rápida.

    1.5. Los arrays son del tipo numpy.ndarray. El método .dtype (**sin paréntesis**) devuelve el tipo de los elementos del array.

    1.6. Se pueden localizar, sobreescribir y hacer slicing de los valores de un array usando los corchetes, como es habitual con otros objetos que contienen varios elementos.

    1.7. El método **.size (sin paréntesis)** te devuelve el número de elementos de un array y el método **.ndim (sin paréntesis)** te devuelve su número de dimensiones (1 para 1D). A su vez, **.shape (sin paréntesis)** te devuelve el número filas, columnas y demás (esto último en caso de un array de 2 o más dimensiones).

    **1.8. Los métodos .mean(), .máx(), .min() y .std()** te devuelven la media, el mínimo, el máximo y la desviación estándar de un array.

    1.9. La suma (o resta), multiplicación (o división) y exponenciación (\*\*) de dos arrays se realiza **elemento a elemento**.

    1.10. El producto escalar se obtiene mediante la función np.dot() introduciendo los dos arrays a multiplicar en el paréntesis, separados por comas.

    1.11. Si usas operadores de comparación sobre un array, obtienes otro array con Trues y Falses según si cada elemento del array cumple o no la condición escrita previamente. Por el contrario, **si escribes array_name[{condición}] lo que obtienes es un nuevo array solo con los elementos que cumplen la condición.**

    1.12. np.pi te devuelve el número pi. También hay otras funciones matemáticas como np.sin() para el seno.

    **1.13. La función np.linspace(x, y, n) crea un array equiespaciado de n elementos, empezando por el valor x y acabando en el valor y (ambos son incluidos en el array).**

2. Numpy - Arrays 2D 

    **2.1. Para crear un array de 2D (una matriz) creamos una lista con listas en su interior y luego le pasamos la función np.array** para castearla al formato array de np. Se puede observar que np la muestra reestructurada con una forma similar a la de una matriz.

    2.2. Los métodos .size, .ndim y .shape siguen funcionando, lógicamente.

    2.3. La **función np.zeros([n,m])** te crea una matriz de ceros con n filas y m columnas. Se puede usar para crear arrays de N dimensiones y, si fuera uno de 1D, no hay que poner el número entre corchetes. **Otra función análoga es np.ones().**

    2.4.np.eye(n) te crea una matriz identidad de dimensiones nxn.

    **2.5. np.arange() es como range, pero para arrays (limitado, por tanto, a 1D).**

    **2.6. La función np.random.rand() te crea un array de N dimensiones, pero cuyos elementos son números aleatorios entre 0 y 1.** Al contrario que en np.zeros(), **aquí no hay que poner los números entre corchetes** si el array tiene más de una dimensión.

    2.7. Para localizar un elemento en una matriz, se puede usar corchetes y los índices, como viene siendo habitual. No obstante, se puede usar una segunda notación donde en lugar de usar un corchete para todas las dimensiones, se usa uno para cada dimensión. El uso de una notación u otra es simplemente cuestión de gustos.
    
    2.8. **Las operaciones básicas** de suma (y resta), multiplicación (división) y exponenciación de matrices, **se aplican elemento a elemento por defecto.**
    
    2.9. Para hacer una multiplicación de matrices habitual, usamos la función np.dot({matrix1}, {matrix2}) o, directamente, usamos {matrix1}@{matrix2}, donde {matrix1} y {matrix2} son las matrices a multiplicar.

    **2.10. Para obtener la traspuesta de una matriz usamos el método .T __sin paréntesis__.
    
    **2.11.El método .astype({un tipo de objeto}) te castea los elementos de un array a {un tipo de objeto} (siempre que dicho cambio sea viable).**
    
    **2.12. El método .reshape({nrows}, {ncols}) te redimensiona un array para que tenga el número de filas y columnas que le hayas indicado en el paréntesis (siempre que dicho cambio sea viable).** Con esta función puedes redimensionar a más de 2D si quieres, añadiendo más números y, además, **puedes poner -1 en uno de esos parámetros para que ponga en esa dimensión "tantos elementos como sea necesario" para poder redimensionar el array original con el resto de parámetros que le hayas indicado.**
    
    **2.13. El método .ravel() aplana un array, esto es, lo vuelve unidimensional (se ve como un vector fila).**
    
    **2.14. np.NaN es un nan, -np.Inf es un "-infinito". Con la función np.isnan({nparray}) puedes detectar dónde están los nan (devuelve un vector booleano) y con np.isfinite({array}) obtendrás un vector booleano donde los False no son finitos (esto incluye a los nan dentro de los Falses). Es útil usar estas funciones como condición para filtrar los datos que sean nan o infinitos.**
    
    **2.15. Para filtrar, muchas veces es útil la negación (\~ en Python).** Ejemplo: a[~np.isnan(a)], donde a es un np array, devolverá un np array con los elementos de a que no son nan's.
