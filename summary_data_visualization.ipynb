2.  Gráficos. Parte 2 (seaborn).

    2.1. Si importamos rcParams (**from matplotlib import rcParams**), podemos cambiar el tamaño por defecto de las gráficas **rcParams['figure.figsize'] = {n},{m}** en el Notebook. Además, con la **función sns.set_context()** puedes cambiar el tamaño de la fuente y cosas por el estilo (esto último está pte. de COT).
    
    2.2. Para **cargar un df** usamos la función **sns.load_dataset({df_name})**. Si es un df predefinido en python, tenemos que poner su nombre entre comillas.
    
    2.3. Si el df es muy grande, podemos pedir una muestra de datos aleatorios del mismo con {var_name} = **{df_name}.sample({n}, random_state=1234).copy()**, que copia de {df_name} {n} datos aleatorios en {var_name}. El argumento **random_state = 1234 es una semilla** (seed), sirve para que la generación de datos aleatorios siempre sea el mismo cuando se ejecuta por n-ésima vez el código, de modo que los resultados sean reproducibles para otra persona que acceda al código posteriormente.
    
    2.4. La función **sns.scatterplot(x=, y=, data =)** genera un gráfico de dispersión. **Argumentos**: size = para tamaños, style = para formas, hue = para color, hue_order = {una lista con los valores/categorías} para que los colores se ordenen como tú desees y palette = '{nombre de una paleta}' para asignar una paleta de colores concreta.
    
    2.5. Por defecto, la escala de colores para una variable continua es un degradado entre dos colores extremos y, para una variable discreta (categorías, para simplificar), son colores lo más diferentes entre sí que sea posible para esa cantidad de categorías.
    
    2.6. La función **sns.stripplot(x='{col_name}', y='{col_name}', data={df_name})** dibuja un gráfico de dispersión cuyo eje X viene dado por una variable categórica, de tal forma que separa en cierta medida los puntos que aparecerían juntos, para que se aprecie la densidad de los mismos en cada categoría.
    
    2.7. La función **sns.swarmplot()** es análoga en cuanto a argumentos a la anterior, pero separa más claramente los datos. Otra opción útil y análoga en cuanto a argumentos para esto es un gráfico de violin, que se dibuja con **sns.violinplot()**.
    
    2.8. Para dibujar un ***boxplot usamos sns.boxplot()**, con los mismo argumentos que sns.stripplot().
    
    2.9. Para dibujar un gráfico de barras podemos usar **sns.countplot(x = , data =) en lugar de sns.barplot()**, pero obsérvese que en este caso la **variable y no se añade, ya que va a ser la frecuencia** (de ahí lo de "countplot").
    
    2.10. Para dibujar un histograma usamos la **función sns.distplot({df_name}.{col_name})**, obsérvese que hemos pasado la variable a representar como pd Series (en este caso no se puede usar data =, x=; que es lo que hacíamos en gráficos anteriores).
    
    2.11. **Por defecto, el histograma se dibuja con la curva de ajuste y el eje Y no es frecuencia, sino densidad**. Para que el eje Y sea frecuencia y dicha curva de ajuste no aparezca,esto es, **para evitar ambas cosas, podemos añadir el argumento kde = False**.
    
    2.12. Para **cambiar un eje X a escala logarítmica**, tras dibujar una gráfica podemos escribir la siguiente línea: **plt.xscale('log')**.
    
    2.13. __El método .unique() es como DISTINCT en SQL__ y se puede usar para sacar una lista con las categorías que contiene la columna de una variable categórica en un df. **Ejemplo:** list({df_name}['{col_name}'].unique()). Esto permite usar las categorías con facilidad dentro de un bucle for, por ejemplo.
    
    2.14. **Si dibujas varias gráficas en la misma caja del Notebook, irán apareciendo unas sobre otras**. Con un **bucle for**, podemos ir almacenando los nombres de cada variable representada en el eje X en el argumento labels = {col_name_iter_ind}. De esa forma, si **fuera del bucle for (y después del mismo)** escribimos **plt.legend()**, se creará una leyenda donde cada color indicará la variable a la que hace referencia (cada histograma que se ha ido pintando, se pinta de un color distinto al anterior, lógicamente). 
    
    2.15. La **función sns.pairplot({df_name}) crea una matriz de gráficas** donde la diagonal principal viene dada por los histogramas de cada variable (columna) individual y las demás posiciones se representan como gráficos de dispersión de cada posible pareja de variables en el df (por lo tanto, sí, hay 2 scatterplots iguales para cada pareja, una en la diagonal superior y otro en la diagonal inferior). **Puedes añadir argumentos** como hue = para distinguir algo más en la matriz, aunque su objetivo fundamental es **hacer una primera inspección del df y observar correlaciones**.
