2. Strings

    2.1. LLamar a una serie de elementos de un string: var_name[index_pos_i:index_pos_f(no entra)]. O simplemente usar una index_pos para un solo caracter.
    
    2.2. Los índices en Python se numeran del 0 al infinito y del -1 al -inf.
    
    2.3. ::2 significa los elementos separados con un paso 2 (0, 2, 4, 6,...). 0:5:2 significaría "hasta el 5 desde el 0 de 2 en 2", esto es, índices 0,2 y 4.
    
    2.4. Se puede usar + para concatenar strings y * para replicar un mismo string n veces (n*{string}).
    
    2.5. Puedes usar \n dentro de un string para saltar línea y \t para añadir un espacio de tabulador. Asimismo, \\ te devuelve \ al imprimit el string (print), pero eso también se puede arreglar añadiendo r antes de empezar el string (antes de la primera comilla), de tal forma que podamos escribir \ directamente en el string. Todo esto es pensando en hacer un print, es decir, si llamas a la variable directamente no va a funcionar.
    
    2.6. Método .upper() para poner un string guardado en una variable en mayúsculas. Análogo con .lower() para minúsculas y .title() para estilo título.
    
    __2.7. Método .replace({substring a reemplazar}, {string que lo reemplaza}) para reemplazar una parte de un string más grande.__
    
    __2.8. Método .find({substring}) te da el primer índice correspondiente al substring que le pides buscar. Devuelve -1 si no puede encontrar el substring. ¡Cuidado! Es sensible a las mayúsculas.__
    
    __2.9. El método .split() te divide un string en varios atendiendo al separador que le indiques en el paréntesis, por ejemplo ',' para dividirlo por comas.__
    
    __2.10. El método .join() hace justo lo contrario y su sintaxis también es la contraria, esto es, para juntar usando comas sería:__
    
    __','.join(var_name)__


3. Tuplas

    3.1. __Van entre paréntesis__ y pueden contener todo tipo de objetos, incluyendo otras tuplas.
    
    3.2. Puedes sobreescribir lo que hay en los índices de un objeto igualando ese valor al nuevo. Ejemplo: my_tuple[1] = 2 sustituye el elemento de índice 1 en my_tuple por un 2.
    
    __3.3. La función len() te da la longitud de una tupla.__
    
    __3.4. La función sorted() te ordena los elementos de la tupla.__
    
    __3.5. Puedes llamar a los elementos de una tupla dentro de una tupla usando un segundo corchete: my_tuple[2][1]__ donde estaríamos llamando al segundo elemento (índice 1) de una tupla que se encuentra en la tercera posición (índice 2) de my_tuple. Análogamente se pueden añadir más corchetes, ya sea para analizar los elementos de, por ejemplo, un string (porque dé la casualidad de que ese es el elemento que acabamos de sacar con ese código) o porque haya más listas dentro de esa lista.
    
    __3.6. Puedes usar \ dentro de una tupla para continuar escribiéndola en la siguiente línea (es útil si te está quedando larga).__


4. Listas

    __4.1. Son como las tuplas, pero van entre corchetes.__ 
    
    __4.2. Si haces my_list2 = my_list1[0:2], te guarda es una variable una lista con solo esos índices de my_list1.__
    
    __4.3. El método .extend({una lista}) usado sobre otra lista te junta ambas listas, extendiendo la primera. Esto mismo lo puedes hacer también con un + (concatenando) o con el método .append()__, solo que a este último solo le puedes añadir los elementos que van a entrar "tal cual", esto es, si pusieras una lista en el paréntesis metería una lista dentro de la otra lista, no las uniría en el sentido estricto de la palabra.
    
    __4.4. El método .pop() elimina el último elemento de una lista.
    
    __4.5. El método .insert({ndex}, {new element}) añade un elemento nuevo a tu lista con el índice que le has indicado.
    
    __4.6. El método .remove({element of the list}) elimina el elemento indicado de la lista.
    
    __4.7. La función del() te elimina lo que le pongas en el argumento de la función, por ejemplo, un elemento de una lista nombrado como list_name[index_of_the_element].
    
    4.8. El método .split(), mencionado previamente, te crea una lista con los elementos en que ha dividido el string. Por defecto divide la cadena de caracteres por donde hay espacios (esto no se mencionó antes).
    
    __4.9. Para copiar una lista__ (de tal forma que quedan unidas para todos los cambios que se hagan sobre cualquiera de las dos en el resto del notebook), __usamos A=B. Para CLONAR una lista__ (lo que implica que no exista dicha dependencia entre ambas) __lo que se hace es escribir A = B[:]. Una vez hecho esto último, A y B son independientes__, esto es, los cambios efectuados sobre una no afectan a la otra.
    
    
5. Sets

    5.1. Van entre llaves. Es una colección de objetos únicos, no hay valores repetidos.
    
    5.2. Puedes usar la función set() para convertir una lista en un set.
    
    5.3. Con el método .add() puedes añadir elementos a un set. No se pueden añadir elementos duplicados.
    
    5.4. Puedes usar el método .remove() para eliminar un elemento de un set.
    
    5.5. Puedes verificar si un elemento está en el set con la sintaxis {element} in set_name.
    
    5.6. Puedes usar el método .difference({set2}) para ver qué elementos son diferentes entre un set1 y el set2 que escribas en el argumento. Pero solo te da los valores del set1.
    
    5.7. Análogamente, .intersection() te da los valores que se encuentran en ambos sets. Esto también se puede conseguir escribiendo set1 & set2 (operadores lógicos).
    
    5.8. El método .union() te devuelve los valores de ambos sets en un nuevo set, esto es, sin repetir ninguno en el output.
    
    5.9. Los métodos .issuperset() e .issubset() te indican si el set que pones a la izquierda del punto es un superset o un subset del que has puesto en el argumento (esto es, que contiene por completo al otro set o es contenido por completo por el otro set, respectivamente).
    
    
6. Diccionarios

    __6.1. Los diccionarios se escriben entre llaves y cada elemento consta de una clave y un valor__, que pueden ser cualquier tipo de objeto. __La sintaxis viene a ser dicc_name = {clave1 : valor1, ... , claven : valorn}.__
    
    __6.2. En este caso se puede llamar a los valores usando la clave de los mismos, es decir dicc_name[clave1] te devolverá un output con el valor de la clave 1, como si estuvieras llamando a su índice en una lista, set o una tupla.
    
    6.3. Se puede usar el método .get() para obtener el valor de la clave que metamos en el argumento, como ocurriría si no usáramos .get(). 
    
    6.4. Si buscas el valor de una clave que no existe en el diccionario y le pasas un type(), te devuelve NoneType (lógicamente).
    
    __6.5. Puedes usar el método .keys() para obtener las claves de un diccionario y el método .values() para obtener sus valores.
    
    __6.6. Los datos se sobreescriben como de costumbre, pero usando las claves en lugar de los índices habituales.
    
    __6.7. Con la función del() puedes eliminar un valor del diccionario (llamándolo por su clave, como hemos visto) y eliminar tanto el valor como la clave del diccionario.
    
    6.8. También se puede usar la sintaxis clave in dicc_name para saber si existe o no una clave en un diccionario.
    
    __6.9. Puedes usar la función zip para relacionar los elementos de una lista como claves con los elementos de otra lista (que serán los valores). Una vez relacionados, puedes aplicarles la función dict() para crear un diccionarios a partir de ellos.__ Si las listas no tuvieran el mismo número de valores, solo se relacionan los que sí que están en ambas listas (los que se pueden emparejar) y los que están de pico no entrarían en el diccionario; pero no daría error.
